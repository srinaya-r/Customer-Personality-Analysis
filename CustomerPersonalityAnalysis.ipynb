### Generating new features

**A.** Adding a new `Customer_For` feature using the `Dt_Customer` and `Recency` features

```python
def extractCustomerTime(args: Tuple[int, datetime]):
    """
        Extracts the number of active days for a customer (the number of days between the first and latest purchase).
        - `args[0]` represents he number of days (`int`) after the latest purchase.
        - `args[1]` represents the date (`datetime`) of the first purchase.
    """
    daysSince, startDate = args
    recencyDate = datetime.now() - timedelta(days=daysSince)
    return (recencyDate - startDate).days

dataDf[DataColumn.CUSTOMER_FOR] = dataDf[[DataColumn.RECENCY, DataColumn.DT_CUSTOMER]].apply(
    lambda args: extractCustomerTime(tuple(args)),
    axis=1
)
```

**B.** Adding a new `Age` column based on `Year_Birth` column:

```python
dataDf[DataColumn.AGE] = date.today().year - dataDf[DataColumn.YEAR_BIRTH]
```

**C.** Creating a new `Spent` column, which is the sum of `MntWines`, `MntFruits`, `MntMeatProducts`, `MntFishProducts`, `MntSweetProducts` & `MntGoldProds` columns

```python
mntCols = [
    DataColumn.MNTWINES,
    DataColumn.MNTFRUITS,
    DataColumn.MNTMEATPRODUCTS,
    DataColumn.MNTFISHPRODUCTS,
    DataColumn.MNTSWEETPRODUCTS,
    DataColumn.MNTGOLDPRODS
]
dataDf[DataColumn.SPENT] = dataDf[mntCols].sum(axis=1)
```

**D.** Adding a new `Children_Total` column, which is the sum of `Kidhome` and `Teenhome` columns

```python
dataDf[DataColumn.CHILDREN_TOTAL] = dataDf[[DataColumn.KIDHOME, DataColumn.TEENHOME]].sum(axis=1)
```

**F.** Generating a new `Is_Parent` feature, based on the `Children_Total` feature.

```python
dataDf[DataColumn.IS_PARENT] = dataDf[DataColumn.CHILDREN_TOTAL].map(lambda total: int(total > 0))
```

**G.** Generating a new `Living_With` feature, based on the `Marital_Status` feature.

```python
class LivingWithStatus(str, TitlecasedStrEnum):
    """ Possible `Living_With` values. """

    PARTNER = auto()
    ALONE = auto()

def mapMaritalStatusToLivingWith(ms: MaritalStatus):
    if ms == MaritalStatus.MARRIED or ms == MaritalStatus.TOGETHER:
        return LivingWithStatus.PARTNER.value
    return LivingWithStatus.ALONE.value

dataDf[DataColumn.LIVING_WITH] = dataDf[DataColumn.MARITAL_STATUS].map(mapMaritalStatusToLivingWith)
dataDf[[DataColumn.LIVING_WITH, DataColumn.MARITAL_STATUS]].head(n=8)
```

**E.** Generating a new `Family_Size` feature, based on `Children_Total` & `Living_With` features.

```python
def findFamilySize(args: Tuple[int, LivingWithStatus]):
    childrenTotal, livingWith = args
    return childrenTotal + int(livingWith == LivingWithStatus.PARTNER.value) + 1

dataDf[DataColumn.FAMILY_SIZE] = dataDf[[DataColumn.CHILDREN_TOTAL, DataColumn.LIVING_WITH]].apply(
    lambda args: findFamilySize(tuple(args)),
    axis=1
)
dataDf[[DataColumn.FAMILY_SIZE, DataColumn.LIVING_WITH, DataColumn.CHILDREN_TOTAL]].head(n=8)
```

**H.** Generating a new `Age_Group` column, based on the `Age` column.

```python
class AgeGroups:
    """
        Used to generate Age Group strings, such as `'Age1 - Age2`', `'< Age'` or `'> Age'`,
        based on the contents of `AgeGroups.groupLimits`.
    """

    groupLimits = [
        21,
        30,
        40,
        50,
        60,
        70,
        80
    ]
    """ A sorted `List` which contains the boundaries for each age group to be generated. """
    assert groupLimits == sorted(groupLimits)

    @staticmethod
    def getAgeGroupStrings():
        """ Returns a `List` with all the age group strings that can be generated using `AgeGroups.groupLimits`. """
        groups: List[str] = []
        for i, gl in enumerate(AgeGroups.groupLimits):
            if i == 0:
                groups.append(f'<{gl}')
                if len(AgeGroups.groupLimits) > 1:
                    groups.append(f'{gl}-{AgeGroups.groupLimits[i + 1]}')
                continue

            if i < len(AgeGroups.groupLimits) - 1:
                groups.append(f'{gl + 1}-{AgeGroups.groupLimits[i + 1]}')

        groups.append(f'>{AgeGroups.groupLimits[-1]}')
        return groups

    @staticmethod
    def mapAgeToAgeGroup(age: int):
        """ Returns the corresponding age group string for `age`. """
        for i, gl in enumerate(AgeGroups.groupLimits):
            if age <= gl:
                if i > 0:
                    return f'{AgeGroups.groupLimits[i-1] + 1}-{gl}'
                else:
                    return f'<{gl}'

        return f'>{AgeGroups.groupLimits[-1]}'

dataDf[DataColumn.AGE_GROUP] = dataDf[DataColumn.AGE].map(AgeGroups.mapAgeToAgeGroup)
dataDf[[DataColumn.AGE_GROUP, DataColumn.AGE]].head(n=8)
```

We eliminate the outliers using the IQR (Inter Quartile Range) method.

```python
def removeColumnOutliers(df: pd.DataFrame, col: str):
    """ Removes `df` rows with outlier value in column `col`, using IQR method. """

    q1, q3 = runWithNoWarnings(
        lambda: (
            np.percentile(df[col], q=25, interpolation='midpoint'),
            np.percentile(df[col], q=75, interpolation='midpoint')
        )
    )
    iqr = q3 - q1

    floor = q3 - 1.5 * iqr
    ceil = q3 + 1.5 * iqr

    df.drop( df.loc[~df[col].between(left=floor, right=ceil)].index, inplace=True )
    df.reset_index(inplace=True, drop=True)

for col in outlierCols:
    removeColumnOutliers(dataDf, col)
```

**Interpreting a heatmap**

A heatmap visualizes the correlation between each 2 features of the dataset:
- A value closer to **1.0** means there is a **linear (proportional)** trend between the features: when `feature_1` value increases, `feature_2` value tends to increase (at least in some degree). This also explains why the diagonal values are all equal to 1.0 (each feature is linearly realted to itself).
- A value closer to **-1.0** means there is an **inverse proportional** relatioship, therefore `feature_1` value tends to decrease when `feature_2` value increases.
- A value closer to **0** means there is no linear relationship between the 2 features.

```python
class HistogramData(TypedDict):

    col: str
    """ Column title. """
    xlabel: str
    """ x-axis label. """

class HistogramGenerator:
    """ Used to generate histogram. """

    def __init__(self) -> None:
        self.questions: List[HistogramData] = []

    def registerQuestions(self, questions: Iterable[HistogramData]):
        self.questions = list(questions)
        return self

    def registerQuestion(self, q: HistogramData):
        self.questions.append(q)
        return self

    def clearQuestions(self):
        self.questions.clear()
        return self

    @staticmethod
    def displayColumnHistogram(df: pd.DataFrame, col: str, xlabel: str, ylabel: str = '# Customers'):
        """
            Displays the Histogram for column `col` of `df`.
            - `xlabel`: the label for x-axis.
            - `ylabel`: the label for y-axis.
        """

        ax = plt.subplot()
        ax.set_title(col)
        ax.set_xlabel(xlabel)
        ax.set_ylabel(ylabel)
        df[col].hist(ax=ax, grid=True, zorder=3)

    def histoGrams(self, df: pd.DataFrame):
        for q in self.questions:
            self.displayColumnHistogram(df, q['col'], xlabel=q['xlabel'])
            plt.show()
            print("/n")
```

A Scatter plot can visualize the distribution of samples, in 2-D area or 3-D space. It will also allow us to examine generated clusters later.

```python
def create2dScatter(arr: np.ndarray, ax: "plt.Axes | None" = None, colors: "np.ndarray | None" = None) -> plt.Axes:
    """
        Creates a 2-D Scatter Plot for the samples in `arr`.
        - `ax`: If specified, the Scatter Plot will be created using this `Axes` object.
        - `colors`: If specified, each plotted point for `arr` will have the respective index color.

        Returns the `Axes` object used for plotting.
    """

    if arr.shape[1] < 2:
            raise Exception(f'Cannot create 2D scatter for `np.ndarray` of shape {arr.shape}.')

    if not ax:
        ax = plt.subplot()

    ax.scatter(x=arr[:, 0], y=arr[:, 1], c=colors)

    return ax

def create3dScatter(arr: np.ndarray, ax: "Axes3D | None" = None, color: "np.ndarray | None" = None) -> Axes3D:
    """
        Creates a 3-D Scatter Plot for the samples in `arr`.
        - `ax`: If specified, the Scatter Plot will be created using this `Axes3D` object
        - `color`: If specified, the plot points will have this color.

        Returns the `Axes3D` object used for plotting.
    """

    if arr.shape[1] < 3:
        raise Exception(f'Cannot create 3D scatter for `np.ndarray` of shape {arr.shape}.')

    if not ax:
        fig = plt.figure(figsize=(7, 6))
        ax = fig.add_subplot(projection='3d')

    ax.scatter3D(xs=arr[:,0], ys=arr[:,1], zs=arr[:,2], c=color)
    return ax
```

Clustering Implementation

To determine the most suitable number of clusters, we will perform the Elbow method and choose the number of clusters where K-Means SSE starts to stabilize.

```python
class ClustererWrapper:
    """ Provides a common interface for `sklearn` clusterers. """

    supportedClusterer = Type["KMeans | AgglomerativeClustering"]

    def __init__(self, cls: supportedClusterer) -> None:
        self.cls = cls

    def initClusterer(self, nClusters: int):
        if self.cls == KMeans:
            self.__clusterer__ = self.cls(n_clusters=nClusters)
            self.__name__ = 'K-Means'
        elif self.cls == AgglomerativeClustering:
            self.__clusterer__ = self.cls(n_clusters=nClusters, compute_distances=True)
            self.__name__ = 'Agglomerative Clustering'

    def getName(self):
        return self.__name__

    def fit(self, data: np.ndarray):
        runWithNoWarnings(
            lambda: self.__clusterer__.fit(data)
        )

    def fitPredict(self, data: np.ndarray):
        return runWithNoWarnings(
            lambda: self.__clusterer__.fit_predict(data)
        )

    def getSSE(self):
        if self.cls == KMeans:
            return self.__clusterer__.inertia_
        elif self.cls == AgglomerativeClustering:
            raise Exception('Cannot get SSE for `AgglomerativeClustering`')

def performElbowMethod(clusterer: ClustererWrapper, data: np.ndarray, verbose: bool = True):
    """
        Performs the Elbow method for the specified `clusterer` and `data`.
        If `verbose` is `True`, a `tqdm` progress bar will be displayed during the execution.

        Returns a list with the number of clusters used in each loop,
        and a list with the respective SSE for each loop.
    """

    iterRange = range(K_MIN, K_MAX + 1)
    if verbose:
        iterator = tqdm(
            iterRange,
            bar_format='{desc}{bar} {n}/{total} -- Time Elapsed: {elapsed}'
        )
    else:
        iterator = iterRange

    errorByK = []
    for k in iterator:
        clusterer.initClusterer(nClusters=k)
        clusterer.fit(data)
        errorByK.append(clusterer.getSSE())

        if verbose:
            iterator.set_description(f"Fitting {clusterer.getName()} with k = {k:<3}")

    return list(iterRange), errorByK

def demonstrateElbowMethod(clusterer: ClustererWrapper, data: np.ndarray, verbose: bool = True):
    """
        Performs the Elbow method for the specified `clusterer` and `data`.
        After completion, it displays a plot showing the SSE for each number of clusters.

        If `verbose` is `True`, a `tqdm` progress bar will be displayed during the execution.
    """

    kSizes, errors = performElbowMethod(clusterer, data, verbose)
    ax = plt.subplot()
    ax.set_title(f'{clusterer.getName()} SSE by #Clusters')
    ax.set_ylabel('SSE')
    ax.set_ylim(0, max(errors))
    ax.set_xlabel('#Clusters')
    ax.grid(visible=True)
    ax.plot(kSizes, errors, marker='o')
    plt.show()

demonstrateElbowMethod(ClustererWrapper(KMeans), pcaArray, verbose=VERBOSE_CLUSTERING)
```

After choosing a suitable number of clusters to use, we perform clustering using `KMeans` and `AgglomerativeClustering`. We then use multi-color scatter plots to visualize the cluster distribution and locate neighboring clusters.

In each case, we display a 3-D plot, as well as **X-Y** & **Y-Z** 2-D plots, to provide different "angles" of the 3-D plot.

```python
class AxisInfo(TypedDict):
    axis: Literal['X', 'Y', 'Z']
    label: str

def mapIndexToAxisInfo(idx: int) -> AxisInfo:
    if idx == 0:
        return {
            'axis': 'X',
            'label': 'PCA Component 0'
        }
    elif idx == 1:
        return {
            'axis': 'Y',
            'label': 'PCA Component 1'
        }
    elif idx == 2:
        return {
            'axis': 'Z',
            'label': 'PCA Component 2'
        }
    else:
        raise Exception(f'Index {idx} cannot be mapped to an axis')

def extractScatterAxesRGB(ax: plt.Axes, collectionIndex: int) -> np.ndarray:
    """
        Returns the RGB color of the **n-th** `Collection` in `ax`,
        where n is `collectionIndex`.
    """
    rgb = ax.collections[collectionIndex].get_facecolor()[0, :3]
    return np.array([rgb])

def create2dClustersScatter(data: np.ndarray, dataClusters: np.ndarray, clusterColors: Sequence[np.ndarray], ax: plt.Axes):
    """
        Creates a 2-D cluster scatter plot for the 3-dimensional samples in `data`.

        Each sample cluster is determined by the respective index in `dataClusters`,
        while the color of the sample marker is determined by the cluster color
        specified in `clusterColors`.

        This routine ensures that z-index (`zorder`) of each marker in 2-D plots is valid:
        The samples are given to `plt.Axes.scatter` sorted by ascending Z value,
        so that the ones with larger Z value are plotted on top of the others.
    """

    sortedIndices: np.ndarray = data[:, 2].argsort()
    # We get a view of `data` sorted by Z, and the cluster color for each sample in `sortedData`
    sortedData: np.ndarray = data[sortedIndices]
    sortedColors: np.ndarray = np.array([clusterColors[c] for c in dataClusters[sortedIndices]])
    return create2dScatter(sortedData[:, [0, 1]], ax, sortedColors)

def visualizeClusters(data: np.ndarray, dataClusters: np.ndarray, clustererName: str):
    """
        Visualizes the distribution of 3-dimensional samples in `data` in clusters,
        using the respective cluster numbers in `dataClusters`.
        The samples of the same cluster will have the same color,
        and samples of different clusters will have different colors.

        In total, 3 plots will be displayed:
        - A 3-D Scatter Plot for X, Y & Z axes.
        - A 2-D Scatter Plot for X & Y axes.
        - A 2-D Scatter Plot for Y & Z axes.

        If samples in `data` have more than 3 dimensions, only the 3 first will be used.

        Note that each sample will have the same RGB color in all plots,
        and the z-index (`zorder`) of each marker in 2-D plots is valid
        and can be validated by the 3-D plot:
        - In the X-Y plot, the z-index order is based on the Z value of the samples
        (the ones with larger Z value are plotted on top of the others.)
        - In the Y-Z plot, the z-index order is based on the X value of the samples
        (the ones with larger Z value are plotted on top of the others.)
    """

    fig = plt.figure(figsize=(25, 6))
    ax3D: Axes3D = fig.add_subplot(1, 3, 1, projection='3d')
    axXY = fig.add_subplot(1, 3, 2)
    axYZ = fig.add_subplot(1, 3, 3)

    activeClusters = np.unique(dataClusters)
    clusterColors = []

    def scatter3dWrapper(ax3D: Axes3D, indices: List[int]):
        for i, cluster in enumerate(activeClusters):
            clusterSamples: np.ndarray = dataClusters == cluster

            ax3D = create3dScatter(data[:, indices][clusterSamples, :], ax3D)

            # We extract the selected color for this cluster,
            # to use it in the 2-D plots as well.
            color = extractScatterAxesRGB(ax3D, i)
            clusterColors.append(color)
        return ax3D

    def scatter2dWrapper(ax: plt.Axes, indices: List[int]) -> plt.Axes:
        return create2dClustersScatter(data[:, indices], dataClusters, clusterColors, ax)

    # We operate on
    # - An `Axes` object used to create a Scatter Plot on
    # - A list of indices to rearrange the column order
    # - A wrapper for the Scatter Plot creator function
    # - A dimension indicator, used to set the plot title and labels accordingly
    for ax, indices, scatterWrapper, dimensions in (
        # We create the 3-D plot first, to populate `clusterColors`
        # with existing colors to use in 2-D plots
        (ax3D, [0, 1, 2], scatter3dWrapper, 3),
        (axXY, [0, 1, 2], scatter2dWrapper, 2),
        (axYZ, [1, 2, 0], scatter2dWrapper, 2)
    ):
        ax = scatterWrapper(ax, indices)

        for idx, setter in zip(indices[:dimensions], (ax.set_xlabel, ax.set_ylabel)):
            axisInfo = mapIndexToAxisInfo(idx)
            setter(axisInfo['label'])

        ax.set_title(f"{'-'.join(mapIndexToAxisInfo(idx)['axis'] for idx in indices[:dimensions])} Plot")

    ax3D.set_zlabel(mapIndexToAxisInfo(2)['label'])
    printMd(f'- Visualization of **{clustererName}** clusters:')
    plt.show()

for cls in (KMeans, AgglomerativeClustering):
    clusterer = ClustererWrapper(cls)
    clusterer.initClusterer(nClusters=N_CLUSTERS)
    clustererOutput = clusterer.fitPredict(pcaArray)
    visualizeClusters(pcaArray, clustererOutput, clusterer.getName())
```
